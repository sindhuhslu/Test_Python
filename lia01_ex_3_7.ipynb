```python
%matplotlib inline 

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage.transform import rescale
img = mpimg.imread('stinkbug4.JPG')
plt.imshow(img)
```

```python
img = rescale(img, .2)

img = np.delete(img, np.arange(20), axis=0)
img = np.delete(img, np.arange(80,100), axis=0)
img = np.delete(img, np.arange(20), axis=1)
img = np.delete(img, np.arange(80,100), axis=1)

col = np.zeros(img.shape[0] * img.shape[1])

k = 0
for i in np.arange(img.shape[0]):
    for j in np.arange(img.shape[1]):
        col[k] = np.round(.99*img[i,j,:],0)
        k = k + 1
        
image = np.array([np.repeat(np.arange(img.shape[0]),img.shape[1]), np.tile(np.arange(img.shape[1]),img.shape[0])]).T
image[:, 0] = image[:, 0] - 40
image[:, 1] = image[:, 1] - 40

image = image[col==0, :]


plt.plot(image[:, 0],image[:, 1],"o",color="black")
plt.axis('square')
plt.xlim(-40, 40)
plt.ylim(-40, 40)
plt.show()
```

```python
image = image.T
np.shape(image)
```

### b)

```python
A = np.array([[0, 1], [1,0]])
B = np.array([[2, 0], [0, 1]])
C = np.array([[0, -1], [-1, 0]])
```

Yes, it is possible to multiply for example <code style="background:grey;color:black">A@image</code> in this order, since <code style="background:grey;color:black">A</code> has 2 columns and <code style="background:grey;color:black">image</code> has 2 rows. The result is a $(2\times2)(2\times 1081)=(2\times 1081)$ matrix.

### c)

```python
image_A = A @ image
image_B = B @ image
image_C = C @ image
```

```python
plot(image_A)
```

Reflection about the axis $y=x$

```python
plot(image_B)
```

Stretching by factor 2 in $x$-direction.

```python
plot(image_C)
```

Reflection about the axis $y=-x$.

### d)

```python
image_BA = B @ (A @ image)
image_AB = A @ (B @ image)

plot(image_BA)
plot(image_AB)
```

Not the same.

```python
image_CA = C @ (A @ image)
image_AC = A @ (C @ image)

plot(image_CA)
plot(image_AC)
```

In this case, the result is the same.
